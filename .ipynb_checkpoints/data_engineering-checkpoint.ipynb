```python
# Dependencies & Setup
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Files to Load
jan2019 = "JC-201901-citibike-tripdata.csv"
jan2020 = "JC-202001-citibike-tripdata.csv"
jan2021 = "JC-202101-citibike-tripdata.csv"
```

```python
# Read July 2016, 2017 & 2018 Citi Bike Data Files & Store into Pandas DataFrames
jan2019_df = pd.read_csv(jan2019, encoding="ISO-8859-1")
jan2020_df = pd.read_csv(jan2020, encoding="ISO-8859-1")
jan2021_df = pd.read_csv(jan2021, encoding="ISO-8859-1")
```

```python
jan2021_df
```

```python
# Combine All Three July DataFrames from 2016-2018 Into Single DataFrame
combined_jan_df = jan2019_df.append([jan2020_df, jan2021_df], ignore_index=True)
combined_jan_df.head()
```

```python
# Rename Columns
combined_jan_df = combined_jan_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
combined_jan_df.head()
```

```python
# Convert All DateTimes to "%Y-%m-%d %H:%M:%S" Format
combined_jan_df["Start Time"] = pd.to_datetime(combined_jan_df["Start Time"])
combined_jan_df["Stop Time"] = pd.to_datetime(combined_jan_df["Stop Time"])
```
